# Titulos, etiquetas, labels, tama√±os. Agregar valor a las visualizaciones


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
#Colocaremos etiquetas titulos y leyendas a las graficas
fig, axs = plt.subplots(1,2,figsize=(20,10))
axs[0].plot(x,y,'g>:', label='sin(x)' )
axs[0].legend()
axs[0].set_xlabel('Eje X')
axs[0].set_ylabel('Eje Y')
axs[0].set_title('Relaci√≥n X vs Y')

axs[1].plot(y,x,'rD--', label='sin(x)' )
axs[1].legend()
axs[1].set_xlabel('Eje Y')
axs[1].set_ylabel('Eje X')
axs[1].set_title('Relaci√≥n Y vs X')
fig.show()
```

```python
#Sin orientaci√≥n a objetos

plt.plot(x,y, label = '$sin(x)$')
plt.legend(loc = 'lower right')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Este es un titulos')
plt.show()
```
