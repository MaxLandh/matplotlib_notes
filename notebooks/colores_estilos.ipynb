# Colores y Estilos

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,6,11)
y = np.sin(x)
```

```python
#Para conocer los estilos que maneja matplotlib utilizaremos el siguiente script

print(plt.style.available)
```

```python
#Para darle un estilo a las figuras primero colocamos este scripts
plt.style.use('dark_background')
fig, axs = plt.subplots(figsize=(20,10))
axs.plot(x,x+1,'r--')
axs.plot(x,x+2,'b:D')
axs.plot(x,x+3,'gs-.')
axs.plot(x,x+4,'y-8')

#Esto es usando notacin matlab
```

```python

plt.style.use('dark_background')
fig, axs = plt.subplots(figsize=(20,10))
axs.plot(x,x+1, color = 'red', marker = 'o', markersize=10, ls='dashdot')
axs.plot(x,x+2,color='blue', marker = 'D',markersize=10)
axs.plot(x,x+3,color='green', marker = '*', markersize=10)
axs.plot(x,x+4,color='yellow', marker = 's', markersize=10)
fig.show()
```
