# Otros graficos

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

```

```python
df = sns.load_dataset('penguins')
df
```

```python
df = df[df['body_mass_g'].isin(df['body_mass_g'].dropna())]
df
```

## Histogramas

```python
#Nos muestra la frecuencia de los datos, i.e. la distribuciÃ³n ordenando los datos de menor a mayor.
plt.figure(figsize=(20,10))
plt.hist(df['body_mass_g'],bins=50)
plt.show()
```

## Boxplot

```python

fig, axs = plt.subplots(1,4, figsize=(20,10))


iterador = 0
for features in df.iloc[:,2:6]:
    axs[iterador].boxplot(df[features], vert=False, patch_artist=True, notch=True)
    iterador += 1
plt.show()
```

## Scatter plot

```python
x = df.body_mass_g
y = df.flipper_length_mm
area = df.bill_depth_mm
color = df.bill_length_mm
plt.figure(figsize=(20,10))
plt.scatter(x,y,s=50, c=color)
plt.show()
```
