# Jointplot, Pairplot

```python
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset('tips')
tips.head(10)
```

```python
#Jointplot: nos une dos tipos de graficas para un mayor analisis

sns.jointplot(data=tips,x='total_bill', y='tip'); 
```

```python
#agregemos un agrupador
sns.jointplot(data=tips,x='total_bill', y='tip', hue='sex'); 
```

```python
sns.jointplot(data=tips,x='total_bill', y='tip', hue='sex', kind='hist', marginal_ticks=True,
                marginal_kws=dict(bins=25)); 
```

```python
sns.pairplot(data=tips, hue='day', corner=True)
```
