# Subplot

Subplot nos permitira tener mas de un grafico a la vez. Veamos como se hace.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,11)
y = np.square(x)

#Utilizaremos el metodo subplot de pyplot
plt.figure(figsize=(20,10))

#Nos pedira 3 parametros, num_filas, num_columnas y el indice
#En este caso sera una figura que contendra una fila y dos columnas, i.e., dos graficas.
#al indicar en el tercer parametro el numero 1 es que esa es la casilla que vamos a graficar.
plt.subplot(1,2,1) 
plt.plot(x,y, 'r--')
#nos movemos a la otra casilla colocando en el tercer parametro la siguiente casilla
plt.subplot(1,2,2)
plt.hist(x)
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
plt.plot(x,y, 'r--')
plt.plot(y,x,'b:')
plt.subplot(1,2,2)
plt.pie(x)
plt.show()
```
