# Metodo orientado a objetos

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,5,11)
y = x ** 2
```

```python
%matplotlib inline
fig = plt.figure() #definimos un lienzo
axes = fig.add_axes([0.1,0.1,1,1])
axes.plot(x, y, 'g:')
axes2 = fig.add_axes([1.1,0.1,1,1])
axes2.plot(x, y, 'rD:')
fig.show()

#Facilmente podemos crear varias graficas en un lienzo con el metodo add_axes
#donde el objeto lista contendra la posición horizontal, posición vertical, ancho y largo de la grafica en su centro de masa.
```
