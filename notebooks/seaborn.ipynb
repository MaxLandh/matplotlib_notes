# Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme(style='darkgrid')
sns.set_palette('muted')
#sns.set(font="Montserrat", font_scale=2)
plt.figure(figsize=(10,10))
sns.barplot(x=['a','b','c'], y=[10,20,30])
plt.show()
```

```python
#Descargamos el dataset tips
import pandas as pd
tips = sns.load_dataset('tips')
tips = pd.DataFrame(tips)
tips
```

```python
#Usaremos la grafica de distribución y veremos la estructura basica de un script de seaborn

sns.displot(tips,x = 'tip', y = 'total_bill', hue = 'sex')
plt.show()

#Son 4 los principales parametros que pide cualquier grafico de seaborn
#1.  La data
#2. eje x
#3. eje y
#4. hue es como los vamos a agrupar.

#Veremos otros parametros que algunas graficas las contienen
```

```python
sns.displot(tips, x = 'total_bill', hue = 'sex', kind='kde', legend=False)
plt.show()

#kind nos permite cambiar el tipo de grafico a una variante de esta
```

```python
penguin = pd.DataFrame(sns.load_dataset('penguins'))
penguin.sample(5)
```
