# Relacional

```python
import matplotlib.pyplot as plt
import seaborn as sns

tips  = sns.load_dataset('tips')
tips.head(12)
```

```python
#scatter plot

sns.scatterplot(data=tips, x='tip',y='total_bill');
```

```python
#podemos agregar mas dimensiones a nuestro grafico de dispersión
plt.figure(figsize=(10,5))
sns.scatterplot(data=tips, x='tip',y='total_bill', hue='day',style='time', size='size');
plt.legend(loc='center', bbox_to_anchor=(1.2,0.5))
```

```python
#lineplot

sns.lineplot(data=tips, x='tip',y='total_bill', hue='day',style='time', size='size');
plt.legend(loc='center', bbox_to_anchor=(1.2,0.5))
```

```python
#relplot
sns.relplot(data=tips, x='tip',y='total_bill', hue='day',style='time', size='size', col='sex', row='time');
```
