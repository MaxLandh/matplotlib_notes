# Categoricos

```python
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset('tips')
tips.sample(3)
```

```python
#Useamos la grafica categorica de countplot

sns.countplot(data=tips, x='day', hue='sex')
```

```python
#stripplot
plt.figure(figsize=(20,10))
sns.stripplot(data=tips,x='day',y='total_bill', hue='sex', dodge=True)
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.swarmplot(data=tips,x='day',y='total_bill', hue='sex', dodge=True)
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.boxplot(data=tips,x='day',y='total_bill', hue='sex', dodge=True);
```

```python
#violinplot
plt.figure(figsize=(10,10))
sns.violinplot(data=tips,x='day',y='total_bill', hue='sex', split=True)
plt.show()
```

```python
#catplot
#catplot es el comodin de las graficas categoricas, ya que se graficara la grafica 
#mas acorde a los datos

plt.figure(figsize=(10,10))
sns.catplot(data=tips,x='day',y='total_bill', hue='sex',kind='box', col='time')
#contiene mas parametros que nos ayudaran a mostrar graficos mejores
plt.show()
```
